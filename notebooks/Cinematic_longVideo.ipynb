***demo***

```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video

pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()

prompt = "bear drives a snow mobile to the moon"
output_name=f'/content/{vid_name}.mp4'
video_frames = pipe(prompt, num_inference_steps=25).frames
video_path = export_to_video(video_frames,output_video_path=output_name)

print(video_path)
```

**Longer video. take longer time and 16GB VRAM card.**

```python
# customize the prompt and output name for long video
prompt = "Duck is swimming"
long_vid_name = 'demo_result_long_video'
```

```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video

# load pipeline
pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)

# optimize for GPU memory
pipe.enable_model_cpu_offload()
pipe.enable_vae_slicing()

# generate
output_name=f'/content/{long_vid_name}.mp4'
video_frames = pipe(prompt, num_inference_steps=25, num_frames=200).frames

# convent to video
video_path = export_to_video(video_frames,output_video_path=output_name)
print(video_path)
```
